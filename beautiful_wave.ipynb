```python
import numpy as np
import matplotlib.pyplot as plt

figure(num=None, figsize=(12, 9), dpi=80, facecolor='w', edgecolor='k')

colors = ['b', 'g', 'y', 'k', 'r']

t=np.arange(-10., 10., 0.02)

for k in range(5):
    line=plt.plot(t,(1/(k+1))*np.sin((k+1)*t))
    plt.setp(line,color=colors[k], linewidth=1.0)
plt.show()
```
